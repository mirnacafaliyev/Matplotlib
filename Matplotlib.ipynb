```python
def decorate(ax):
    # set range for x and y axis
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.set_xticks([])
    ax.set_yticks([])

def sad_emoji(ax):
    plt.title('Keep Smile, Work Hard!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 25})

    ax.scatter([.5],[.5], c='#9dcdf8', s=120000, label="face")
    ax.scatter([.35, .65], [.63, .63], c='k', s=1000, label="eyes")

    x = np.linspace(.3, .7, 50)
    y = -(2*(x-.5)**2 - .4)

    ax.plot(x, y, c='k', linewidth=8, label="smile")

    decorate(ax1)

def smile_emoji(ax=None):
    plt.title('Success is not far!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 25})
    
    ax.scatter([.5],[.5], c='#f6fa0e', s=120000, label="face")
    ax.scatter([.35, .65], [.63, .63], c='k', s=1000, label="eyes")

    X = np.linspace(.3, .7, 50)
    Y = (2*(X-.5)**2 + .32)

    ax.plot(X, Y, c='k', linewidth=8, label="smile")
    
    decorate(ax2)


# Plot in the same line
fig = plt.figure(figsize = (14,14))


ax1 = fig.add_subplot(121, aspect=1)      # ax = fig.add_subplot(1,1,1, aspect=1) # equivalent
sad_emoji(ax1)

ax2 = fig.add_subplot(122, aspect=1)
smile_emoji(ax2)
```

### Bar/Barh

```python
fifa = pd.read_csv('fifa_data.csv')
print(fifa.shape)
fifa.head(3)
```

```python
top_10_country = fifa['Nationality'].value_counts()[:10]
```

```python
plt.bar(top_10_country.keys(), top_10_country, color='#b9a1f0')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.barh(top_10_country.keys(), top_10_country, color='#93fd97')
plt.show()
```

### Scatter

```python
fifa['Weight_in_kg'] = [int(w.strip('lbs'))*.45 if type(w)==str else w*.45 for w in fifa['Weight']]
```

```python
fifa['Height_in_cm'] = [float(h.replace("'", '.'))*30.48 if type(h)==str else h*30.48 for h in fifa['Height']]
```

```python
plt.scatter(data= fifa, x= 'Weight_in_kg', y= 'Height_in_cm', s=8, alpha=.7, c= '#37a0ef')

plt.show()
```

### Boxplot

```python
plt.figure(figsize=(6,6))

barcelona = fifa[fifa['Club']=='FC Barcelona']['Overall']
mancity = fifa[fifa['Club']=='Manchester City']['Overall']
galatasaray = fifa[fifa['Club']=='Galatasaray SK']['Overall']
data = [barcelona, mancity, galatasaray]

labels = ['Barca', 'Mancity', 'Galatasaray']
bp = plt.boxplot(data, labels = labels, notch=True, patch_artist=True, 
                 medianprops={'linewidth': 3, 'color':'k'})

for box in bp['boxes']:
    # change outline color
    box.set(color='#604ff0', linewidth=2)
    # change fill color
    box.set(facecolor = '#79f0ec' )
    # change hatch
    box.set(hatch = '.')

plt.title('Professional Soccer Team Comparison', fontdict={ 'fontsize': 14})
plt.ylabel('FIFA Overall Rating', fontdict={ 'fontsize': 14})

plt.show()
```

### Pie

```python
left_foot = fifa[fifa['Preferred Foot'] == 'Left'].count()[0]
right_foot = fifa[fifa['Preferred Foot'] == 'Right'].count()[0]

data = [left_foot, right_foot]
labels = ['LEFT_FOOT', 'RIGHT_FOOT']
colors = ['#B965fF', '#5bcdef']
explode = [.02,.02]

plt.pie(data, labels=labels, colors=colors, explode=explode, autopct='%.2f %%')

plt.title('FOOT PREFERENCE')

plt.show()
```

### Donut Chart

```python
plt.hist(fifa['Weight_in_kg'], bins=15, color='#9bc3e7')
plt.show()
```

```python
light = fifa[fifa['Weight_in_kg'] < 60].count()[0]
light_medium = fifa[(fifa['Weight_in_kg'] >= 60) & (fifa['Weight_in_kg'] < 70)].count()[0]
medium = fifa[(fifa['Weight_in_kg'] >= 70) & (fifa['Weight_in_kg'] < 80)].count()[0]
heavy_medium = fifa[(fifa['Weight_in_kg'] >= 80) & (fifa['Weight_in_kg'] < 90)].count()[0]
heavy = fifa[fifa['Weight_in_kg'] >= 90].count()[0]

data=[light, light_medium, medium, heavy_medium, heavy]
labels = ['LIGHT', 'LIGHT_MEDIUM', 'MEDIUM', 'HEAVY_MEDIUM', 'HEAVY']
explode = [.01,.01,.01,.01,.02]

plt.pie(data, explode= explode, autopct= '%.f %%',  
        radius=1, wedgeprops=dict(width=.3, edgecolor='w'))


plt.legend(labels, bbox_to_anchor=(1.1,.9), fancybox= True, shadow= True)
plt.title('WEIGHT CATEGORIES')

plt.show()
```

### Violinplot

```python
plt.figure(figsize=(10,6))

barcelona = fifa[fifa['Club']=='FC Barcelona']['Overall']
mancity = fifa[fifa['Club']=='Manchester City']['Overall']
galatasaray = fifa[fifa['Club']=='Galatasaray SK']['Overall']
data = [barcelona, mancity, galatasaray]

labels = ['Barca', 'Mancity', 'Galatasaray']
vp = plt.violinplot(data, showmeans=True )


for v in vp['bodies']:
    # change outline color
    v.set(color='#f20b7a', linewidth=2)
    # change fill color
    v.set(facecolor = '#0def15' )

plt.title('Professional Soccer Team Comparison', fontdict={ 'fontsize': 14})
plt.ylabel('FIFA Overall Rating', fontdict={ 'fontsize': 14})

plt.show()
```
